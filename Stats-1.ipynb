```python
# Solution for problem 1
import math
marks = [6,7,5,7,7,8,7,6,9,7,4,10,6,8,8,9,5,6,4,8]
def mean(marks):
  return sum(marks)/len(marks)
def median(marks):
  if len(marks)%2==0:
    return (marks[len(marks)//2]+marks[(len(marks)//2)+1])//2
  else:
    return marks[len(marks)//2]
def mode(marks):
  dict_ = {}
  for i in marks:
    if i not in dict_:
      dict_[i] = 1
    else:
      dict_[i] += 1
  return max(zip(dict_.values(), dict_.keys()))[1]

def variance(list_):
  mean_ = mean(list_)
  deviations = [(x - mean_) ** 2 for x in list_] 
  variance = sum(deviations) / len(list_)
  return variance

def std(marks):
  var = variance(marks)
  return math.sqrt(var)
```

```python
print(f'mean = {mean(marks)}, median = {median(marks)}, mode = {mode(marks)}, standard deviation = {std(marks)}')
```

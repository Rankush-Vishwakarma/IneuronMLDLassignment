```python
import numpy as np
import math
import scipy.stats as stats
from scipy.stats import chi2_contingency
from scipy.stats import norm
# we have
population_mean = 52
standard_deviation = 4.50
sample_no = 100
sample_mean = 52.80
alpha = 0.05 # 5%

# since the student select random sample from the population data we neeed to measure the standard error 
# it helps us to estimate how well your sample data represents the whole population.
standard_error = standard_deviation/ math.sqrt(sample_no)
# calculating the Z score value
Z_score  = round((sample_mean-population_mean)/standard_error,2)
lowerRegion, upperRegion  = round(norm.ppf(alpha/2),2), round(-norm.ppf(alpha/2),2)
print(f'Z_score : {Z_score}')
print(f'lowerRegion: {lowerRegion}, upperRegion: {upperRegion}')
```

**From above we can conclude we can accept null hypothesis since z_score<critical region**

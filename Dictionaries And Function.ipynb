```python
class Side_of_triangle(object):
  def __init__(self,a,b,c):
    self.length = a
    self.breadth = b
    self.height = c
```

```python
class area_of_triangle(Side_of_triangle):
  def calculate_area(self):
    perimeter = (self.length + self.breadth + self.height)/2
    return (perimeter*(perimeter-self.length)*(perimeter-self.breadth )*(perimeter-self.height)) ** 0.5

```

```python
area = area_of_triangle(3,2,4)
```

```python
area.calculate_area()
```

```python
def filter_long_words(word_lst, n):
  word_len_dict = {}
  for i in word_lst:
    word_len_dict[i] = len(i)
  return [key for key,value in word_len_dict.items() if value > n]
```

```python
filter_long_words(['hi','ineuron','team','thankyou ','for','helping','student','to','build','their','life'],2)
```

```python
def map_list(seq):
  return [len(i) for i in seq]
```

```python
def check_vowel(ch):
  if ch in ['a','e','i','o','u']:
    return True
  else:
    return False
```

```python
check_vowel('a'),check_vowel('b'),check_vowel('z')
```

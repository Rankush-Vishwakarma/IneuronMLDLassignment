```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df.info()
```

```python
df['FlightNumber'] = pd.Series(df.FlightNumber[0]+i for i in range(0,10*(len(df)+1),10)).astype(np.int64)
```

```python
df.info()
```

```python
df
```

```python
temp_dataframe = pd.DataFrame(columns = ['Fname','Lname'])
```

```python
temp_dataframe[['Fname','Lname']] = df.From_To.str.split("_",expand=True)
```

```python
temp_dataframe
```

```python
temp_dataframe['Fname'] = temp_dataframe.Fname.str.lower()
temp_dataframe['Lname'] = temp_dataframe.Lname.str.lower()
```

```python
temp_dataframe
```

```python
df = df.drop('From_To',axis=1)
```

```python
df
```

```python
df.insert(0, "Fname",temp_dataframe.Fname, True)
df.insert(1, "Lname",temp_dataframe.Lname, True)
```

```python
df
```

```python
df = df.RecentDelays.apply(pd.Series).merge(df, right_index = True, left_index = True)
```

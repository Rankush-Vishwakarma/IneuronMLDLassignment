```python
def myreduce(func, seq):
  num0 = seq[0]
  for i in range(1,len(seq)):
    num0 = func(num0, seq[i]) # at initial the function will start from initial two values and storing into variable 
  return num0   
```

```python
def multiply(a,b):
  return a*b
```

```python
myreduce(multiply,[1,2,3,4,5])
```

```python
myreduce(lambda x,y: x+y, [1,2,3,4,5])
```

```python
def myfilter(func, seq):
  new_seq = []
  for i in range(len(seq)):
    if func(seq[i]):
      new_seq.append(seq[i])
  return new_seq
```

```python
def filter_even(num):
    if num%2 == 0:
      return True
```

```python
myfilter(filter_even,[1,2,3,4,5,6])
```

```python
def filter_odd(num):
  if num%2 != 0:
    return True
```

```python
myfilter(filter_odd,[1,2,3,4,5,6])
```
